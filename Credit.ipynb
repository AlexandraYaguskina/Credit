```python
#Переименуем столбцы
df=df.rename(columns={'children':'number_of_children', 'dob_years':'client_age', 'income_type':'employment_type', 'debt':'credit_status', 'purpose':'credit_purpose'})
```

```python
#Подсчет количества пропусков
df.isnull().sum()
```

```python
#Замена пропусков на 0
df=df.fillna(0)
```

```python
#Проекрка на содержание дубликатов и их удаление
df[df.duplicated()].count()
```

```python
df=df.drop_duplicates(keep='first')
```

```python
#Посмотрим на столбец с детьми
df['number_of_children'].unique()
```

```python
#Считаем, что -1-ошибка и заменим на 1; 20 заменим на 2
df['number_of_children']=df['number_of_children'].replace([-1, 20],[1,2])
```

```python
#Уберем отрицательный трудовой стаж с помощью abs
df['days_employed']=df['days_employed'].abs()
```

```python
df['days_employed'].describe()
```

```python
#Максимальное количество дней 401755 - нереально столько проработать, можно по максимальному возрасту вычислить, какое количество дней стажа будет действительно максимальным
df['client_age'].max()
```

```python
#Предположим он работал с 16 лет
max_days_employed=(df['client_age'].max()-16)*365
max_days_employed
```

```python
#Посмотрим, сколько человек превышает значение max_days_employed
df[df['days_employed']>max_days_employed].count()
```

```python
#Переведем стаж из дней в год и изменим тип столбца
df['years_employed']=df['days_employed']/365
df['years_employed']=df['years_employed'].astype(int)
del df['days_employed']
```

```python
#Переведем строки образования в нижний регистр
df.education=df.education.str.lower()
```

```python
#Как разные цели кредита влияют на его возврат в срок?
```

```python
#Нужно по ключевым словам разобрать причины на основные категории: образование, недвижимость, автомобиль и свадьба
def categorize_purpose(credit_purpose):
    if 'свадьб' in credit_purpose:
        return 'Свадьба'
    elif 'образован' in credit_purpose:
        return 'Образование'
    elif 'недвижим' in credit_purpose or 'жиль' in credit_purpose:
        return 'Недвижимость'
    elif 'автомоб' in credit_purpose:
        return 'Автомобиль'
    else:
        return 'Другое'
df['purpose_category'] = df['credit_purpose'].apply(categorize_purpose)
```

```python
df
```

```python
#Отдают кредит вовремя
purpose_credit_0=df[df['credit_status']==0].groupby('purpose_category')['credit_status'].count()
```

```python
#Отдают кредит невовремя
purpose_credit_1=df[df['credit_status']==1].groupby('purpose_category')['credit_status'].count()
```

```python
purpose_credit=df.groupby('purpose_category')['credit_status'].count()
```

```python
purpose_perc= pd.concat([purpose_credit_0, purpose_credit_1], axis= 1)
purpose_perc.columns=['credit_status_0', 'credit_status_1']
```

```python
purpose_perc
```

```python
purpose_perc['perc']=purpose_perc['credit_status_0']*100/purpose_credit
```

```python
purpose_perc
```

```python
#Вывод - зависимость от цели получения кредита наблюдается. Лучше всего возвращают кредит вовремя в категории Недвижимость
#Общий вывод - люди без детей, берущие кредит на недвижимость, имеющие очень высокую заплату и при этом вдовцы, являются наимолее добросовестными заемщиками
```
